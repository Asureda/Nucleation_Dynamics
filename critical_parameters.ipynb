```python
def conversion_pint(K,T,convert):
    sol = []
    if convert:
        for kid, k in enumerate(K):
            sol_ = k.to('Pa**0.5')
            sol_ *= 6.941 #to g
            sol_*= 1/Li_properties.density(T[kid]).magnitude #to cm3 Li
            sol_ *= 1e-6
            #k /= (6.02214076*1e23)
            sol_f = 1/sol_
            #k *= 6.02214076*1e23
            sol.append(sol_f.magnitude)
        return sol
```

```python
# Function to interpolate and add more points
def interpolate_and_add_points(dataframe, num_points=100):
    new_dfs = []
    for i in range(len(dataframe) - 1):
        x_values = np.linspace(dataframe.iloc[i]['X'], dataframe.iloc[i + 1]['X'], num=num_points, endpoint=False)
        y_values = np.linspace(dataframe.iloc[i]['Y'], dataframe.iloc[i + 1]['Y'], num=num_points, endpoint=False)
        temp_df = pd.DataFrame({'X': x_values[1:], 'Y': y_values[1:]})  # Exclude the first point to avoid duplication
        new_dfs.append(dataframe.iloc[[i]])
        new_dfs.append(temp_df)
    new_dfs.append(dataframe.iloc[[-1]])  # Add the last original point
    return pd.concat(new_dfs).reset_index(drop=True)

# Add more points
df_interpolated = interpolate_and_add_points(df)
```
